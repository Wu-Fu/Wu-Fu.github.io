# 图像增广

在CNN中，我们看到了模型在大数据的驱动下展示出的强大性能，但无论如何，我们所能收集到的数据都是有限的，为了保障CNN模型数据的丰富性，我们将从输入端对数据进行操作，来扩大化已有的数据，而且随机改变样本有助于减少模型对数据的依赖，从而提高模型的泛化能力

而这些操作的理论可行性可见于CNN的**平移不变性**

常见的图像增广方法有
- 翻转
- 裁剪
- 过滤和锐化
- 模糊
- 旋转，平移，剪切，缩放
- 剪下
- 色彩
- 亮度
- 对比
- 均匀和高斯噪声
- 渐变
- 镜头变形

[pytorch-torchvision.transforms](https://pytorch.org/vision/stable/transforms.html)

```python
%matplotlib inline
import torch
import torchvision
from torchvision import transforms
import matplotlib
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = 'cat5.jpg'
img = Image.open(image_path)
plt.imshow(img)
```

### 翻转

transforms.RandomHorizontalFlip(p=0.5)

以给定的概率随机*水平翻转*给定的图像

p float 表示给定的概率

返回PIL或Tensor

[pytorch](https://pytorch.org/vision/stable/generated/torchvision.transforms.RandomHorizontalFlip.html#torchvision.transforms.RandomHorizontalFlip)

```python
img1 = transforms.RandomHorizontalFlip(p=1)(img)
plt.imshow(img1)
```
